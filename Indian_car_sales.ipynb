```python
# How many unique car brands are there?
df["Brand"].unique()
```

```python
# Show total cars sold by each brand
df["Brand"].value_counts()
```

```python
# What is the average price of all cars?
df["Price"].mean()
```

```python
#  Find the top 5 most expensive cars
df.sort_values("Price",ascending=False).head(5)
```

```python
# Filter all cars with Price above ₹10 lakhs
df[df["Price"] > 1000000]
```

```python
# Plot number of cars sold by brand
import matplotlib.pyplot as plt
import seaborn as sns

sns.countplot(data=df, x="Brand", order=df["Brand"].value_counts().index)
plt.xticks(rotation=45)
plt.title("Cars Sold by Brand")
plt.show()
```
